```python
import matplotlib.pyplot as plt
from matplotlib.offsetbox import OffsetImage, AnnotationBbox

# This below code is for implementing images in a grid.
# First of all, we have to set figure size, then use a function for reading images from the paths which we define.
# Then there is an additional functionality for zooming of the images.
# Then we use markers for implementing labels of pit, breeze etc which are in wumpus world grid.

plt.rcParams["figure.figsize"] = [7.5, 5.5]
plt.rcParams["figure.autolayout"] = True

def getImage(path,z1):
    return OffsetImage(plt.imread(path, format="jpg"), zoom=z1)

paths = ['Cute-Monster.png', 'gold image 2.jpg', 'Agent.png']
x = [1, 2, 1]
y = [3, 3, 1]

z = [0.06,0.09,0.06]
fig, ax = plt.subplots()

for x0, y0, path, zoom in zip(x, y, paths,z):
    ab = AnnotationBbox(getImage(path,zoom), (x0-.5, y0-.5), frameon=False)
    ax.add_artist(ab)
plt.grid(linestyle="-", color='b')

# Now the markers code begins: 
x = 3
y = 3

print("the value of x is  ", x)
print("the value of y is  ", y)

x,y = ( 2*x + 1)/2,( 2*y + 1)/2
plt.plot(x-1, y-1, color = 'orange',marker = "$ Pit $",markersize=40)
plt.plot(x-1, y-3, color = 'b',marker = "$Pit$",markersize=40)
plt.plot(x, y, color = 'g',marker = "$Pit$",markersize=40)

plt.plot(x-2, y-3, color = 'lightblue',marker = "$Breeze$",markersize=50)
plt.plot(x, y-3, color = 'lightblue',marker = "$Breeze$",markersize=50)
plt.plot(x-1, y-2, color = 'lightblue',marker = "$Breeze$",markersize=50)
plt.plot(x-1, y, color = 'lightblue',marker = "$Breeze$",markersize=60)
plt.plot(x-1, y, color = 'lightblue',marker = "$Breeze$",markersize=60)
plt.plot(x, y-1, color = 'lightblue',marker = "$Breeze$",markersize=60)

plt.plot(x-3, y-2, color = 'g',marker = "$Stench$",markersize=60)
plt.plot(x-3, y, color = 'g',marker = "$Stench$",markersize=60)


# Now the formatting of the grid begins:
plt.xticks(range(0,5))
plt.yticks(range(0,5))
plt.title("Wumpus World 4x4 Grid", color="black", fontsize="35")
plt.xlabel("x-Values", color="Green", fontsize="30")
plt.ylabel("y-Values", color="Blue", fontsize="30")

plt.show()
```
