```python
%matplotlib inline
import numpy as np
x = np.linspace(0, 10, 50)
plt.figure()  # create a plot figure

# create the first of two panels and set current axis

plt.subplot(2, 1, 2) # (rows, columns, panel number)
plt.plot(x, np.sin(x))

# create the second panel and set current axis

plt.subplot(2, 1, 1)
plt.plot(x, np.cos(x));
```

```python
fig = plt.figure()
ax = plt.axes()
x = np.linspace(0,10,50) 

# linspace Parameters: 
# start  : [optional] start of interval range. By default start = 0
# -> stop   : end of interval range
# -> restep : If True, return (samples, step). By default restep = False
# -> num    : [int, optional] No. of samples to generate
# -> dtype  : type of output array

plt.plot(x, np.sin(x - 0), color='blue')        # specify color by name
plt.plot(x, np.sin(x - 1), color='brown')           # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75')        # Grayscale between 0 and 1
plt.plot(x, np.sin(x - 3), color='#FFDD44')     # Hex code (RRGGBB from 00 to FF)
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 and 1
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python
# For Linestyle, we use the following command:
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');

 # For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-')  # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':');  # dotted
```

```python
# If you would like to be extremely terse, these linestyle and color codes can be combined into a single
# nonkeyword argument to the plt.plot() function.

plt.plot(x, x+0 , '-r'); # solid red
plt.plot(x, x + 1, '--g') # dashed green
plt.plot(x, x + 2, '-.b') # dashdot blue
plt.plot(x, x + 3, ':k');  # dotted black
```

```python
# Matplotlib does a decent job of choosing default axes limits for your plot, but sometimes it’s nice to have
# finer control. The most basic way to adjust axis limits is to use the plt.xlim() and plt.ylim() methods .

plt.plot(x,np.sin(x))
```

```python
plt.plot(x,np.sin(x))
plt.xlim(-1,12) # so it is adjusting the figure or u can ssay that plot or grid between the numbers which we give.
plt.ylim(-1.5,1.5)
```

```python
# A useful related method is plt.axis() (note here the potential confusion between axes with an e, and axis with an i).
# The plt.axis() method allows you to set the x and y limits with a single call, by passing a list that specifies 
# [xmin, xmax, ymin, ymax] .
plt.plot(x, np.sin(x))
plt.axis([0.10,-1.5,1.5]) # Must remember, we have to pass a list b/c the plt.axis() method takes only one argument.
```

```python
fig = plt.figure()
fig.subplots_adjust(hspace=0.6, wspace=1.0) # wspace = width-space, hspace = height-space
for i in range(1, 7):
    ax = fig.add_subplot(2, 3, i)
    ax.text(0.5, 0.5, str((2, 3, i)),
            fontsize=18, ha='center') # this is for adding a text into the center.
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')

plt.legend();
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array([80, 90])
y = np.array([240, 250])

plt.title("Sports Watch Data", fontsize=30, color="blue")
plt.xlabel("Average Pulse", fontsize=20, color="black")
plt.ylabel("Calorie Burnage", fontsize=20, color="red")
plt.xticks(range(80,130))
plt.yticks(range(240,300))
plt.grid(linestyle='-')
plt.figsize = (4,4)
plt.plot(x, y)

plt.grid(color = 'green', linestyle = '-', linewidth = 1.0,)

plt.show()
```
